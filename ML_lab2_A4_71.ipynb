```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
df.describe()
```

```python
cols = df.shape[1]
```

```python
x = df.iloc[:,0:cols-1]
y = df.iloc[:,cols-1:cols]
```

```python
x.head()
```

```python
y.head()
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[3300]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
3300*135.78767123 + 180616.43835616432
```

```python
model.predict([[5000]])
```

```python
area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
p = model.predict(area_df)
p
```

```python
area_df['prices']=p
area_df
```
